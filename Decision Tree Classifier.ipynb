```python

import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import io

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

df = pd.read_csv('Book4.csv')
print(df)

d = {"Oki lukman": 0, "Pasha ungu": 1, "Budi anduk": 2}
df['NAMA PASIEN'] = df['NAMA PASIEN'].map(d)

d = {"Overweight": 0, "Average": 1, "Underweight": 2}
df['BERAT BADAN'] = df['BERAT BADAN'].map(d)

d = {'Tua': 0, 'Muda': 1}
df['USIA'] = df['USIA'].map(d)

d = {"Perempuan": 0, "Laki-laki": 1}
df['JENIS KELAMIN'] = df['JENIS KELAMIN'].map(d)

d = {'Ya': 1, 'Tidak': 0}
df['HIPERTENSI'] = df['HIPERTENSI'].map(d)

features = ['BERAT BADAN', 'USIA', 'JENIS KELAMIN']

X = df[features]
y = df['HIPERTENSI']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

plt.figure(figsize=(10, 10))
tree.plot_tree(dtree, feature_names=features)

# Menggunakan io.BytesIO() untuk menyimpan output plot
buf = io.BytesIO()
plt.savefig(buf, format='png')
buf.seek(0)

# Menuliskan output plot ke sys.stdout.buffer
#sys.stdout.buffer.write(buf.getvalue())
#sys.stdout.flush()

plt.show()
```
